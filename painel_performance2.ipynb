```python
import pandas as pd

# --- ETAPA 1: Carregar o arquivo ---
df_propostas = pd.read_csv("propostas.csv", sep=";", encoding="utf-8", low_memory=False)
df_cadastro = pd.read_excel("de_para_loja.xlsx")
df_canal =  pd.read_excel("de_para_canal.xlsx")

df_propostas_2 = df_propostas.copy()

# --- ETAPA 2: Preparar Colunas ---

# Renomenado coluna 'Status' de df_cadastro
df_cadastro.rename(columns={'Status': 'Status_loja'}, inplace=True)

# Renomear coluna 'qualificado' de df_propostas_2
df_propostas_2.rename(columns={'qualificado': 'Qualificado'}, inplace=True)

# --- ETAPA  : Fazer a junção

# Merge de df_propostas_2 e df_canal por 'Canal'
df_propostas_2 = pd.merge(df_propostas_2, df_canal, on='Canal', how='left')

# Merge de df_propostas_2 e df_cadastro por 'Nome PDV'
df_propostas_2 = pd.merge(df_propostas_2, df_cadastro, on='Nome PDV', how='left', indicator=True)



# Criar arquivo .csv
df_propostas_2.to_csv("teste.csv",sep=";" ,index=False)

df_propostas_2.head()
```

```python
# criar dicionario para df_propostas_2 'Canal' de df_canal 'Canal'
dic_canal = dict(zip(df_canal['Canal'], df_canal['nominal']))

# alterar conforme dic_canal
df_propostas_2['Canal'] = df_propostas_2['Canal'].map(dic_canal)
```
