# Relatório Plataformas

```python
import pandas as pd

# --- ETAPA 1: Carregar o arquivo ---
df_propostas = pd.read_csv("propostas.csv", sep=";", encoding="utf-8", low_memory=False)
df_cadastro = pd.read_excel("de_para_loja.xlsx")
df_canal =  pd.read_excel("de_para_canal.xlsx")

df_propostas_2 = df_propostas.copy()

# --- ETAPA 2: Preparar Colunas ---

# Renomenado coluna 'Status' de df_cadastro
df_cadastro.rename(columns={'Status': 'Status_loja'}, inplace=True)

# Trocar na coluna 'Origem' onde for 'MobiBot' por 'Internet' de df_propostas_2
df_propostas_2['Origem'] = df_propostas_2['Origem'].replace('MobiBot', 'Internet')

# Renomear coluna 'qualificado' de df_propostas_2
df_propostas_2.rename(columns={'qualificado': 'Qualificado'}, inplace=True)

# ---  ETAPA : Filtrar Colunas ---

# Filtrar coluna 'Origem' onde for 'Internet' df_propostas_2
df_propostas_2 = df_propostas_2[df_propostas_2['Origem'] == 'Internet']

# Filtrar coluna 'Status' por linhas que sejam diferentes de 'CRM' df_propostas_2
df_propostas_2 = df_propostas_2[df_propostas_2['Status'] != 'CRM']

# --- ETAPA  : Fazer a junção

# Merge de df_propostas_2 e df_canal por 'Canal'
df_propostas_2 = pd.merge(df_propostas_2, df_canal, on='Canal', how='left')

# Merge de df_propostas_2 e df_cadastro por 'Nome PDV'
df_propostas_2 = pd.merge(df_propostas_2, df_cadastro, on='Nome PDV', how='left', indicator=True)

# Filtrar coluna 'Status_loja' onde for 'Ativa' de df_cadastro
df_propostas_2 = df_propostas_2[df_propostas_2['Status_loja'] == 'Ativa']

# Alterar coluna 'Canal' por 'De/Para nominal'
df_propostas_2['Canal'] = df_propostas_2['De/Para nominal']

# Renomear coluna 'De/Para Ajustado' para 'Canal_2'
df_propostas_2.rename(columns={'De/Para Ajustado': 'Canal_2'}, inplace=True)

# --- ETAPA : Fazer agrupamento ---

# Agrupar leads por mês na coluna 'Data' , 'Canal' e 'Qualificado'

leads_por_canal_mes = df_propostas_2.groupby([pd.to_datetime(df_propostas_2["Data"], format="%d/%m/%Y", errors='coerce').dt.to_period("d"),
                                            "Canal","Qualificado"]).size().reset_index(name="Quantidade de Leads")

# --- ETAPA : Criar arquivo csv e download ---

# Criar arquivo .csv
leads_por_canal_mes.to_csv("leads_por_canal_mes.csv",sep=";" ,index=False)

# Download do arquivo
from google.colab import files
files.download("leads_por_canal_mes.csv")

# Vizualização para conferência
leads_por_canal_mes.head()

```
